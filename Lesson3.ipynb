```python
import torch

a = torch.tensor([1,2,3])
print(a, type(a))

print('--------- 2次元のリストから生成 ------------')
b = torch.tensor([
    [1,2],
    [3,4]
])
print(b)

print('--------- dtypeを指定して倍精度のTensorにする ----------')
c = torch.tensor([
    [1,2],
    [3,4],
], dtype=torch.float64)
print(c)

print('---------- 0から9までの数値で初期化 -----------')
d = torch.arange(0, 10)
print(d)

print('---------- 全ての値が0の2 x 3のTensor ----------')
e = torch.zeros(2, 3)
print(e)

print('--------- 全ての値が乱数の2 x 3のTensor ---------')
f = torch.rand(2, 3)
print(f)

print('--------- Tensorの形状はsizeメソッドで取得 --------')
print(f.size())

print('--- -5から5までの連続値を10生成 ---')
g = torch.linspace(-5, 5, 10)
print(g)

print('--- Tensor -> Numpy ---')
h = torch.tensor([
    [1, 2],
    [3, 4],
])
hh = h.numpy()
print(hh)

print('--- Numpy -> Tensor ---')
k = torch.from_numpy(hh)
print(k)

m = torch.tensor([
    [1, 2, 3],
    [4, 5, 6],
])

print('--- 2つのインデックスを指定 ---')
print(m[0, 1])

print('--- 範囲を指定 ---')
print(m[1:2, :2])

print('--- リストで複数のインデックスを指定 ---')
print(m[:, [0, 2]])

print('--- 3より大きい要素のみを指定 ---')
print(m[m > 3])

print('--- 要素を変更 ---')
m[0, 2] = 11
print(m)

print('--- 要素の一括変更 ---')
m[:, 1] = 22
print(m)

print('--- 10より大きい要素のみ変更 ---')
m[m > 10] = 33
print(m)

#********************** 3.2.5 ****************************************
```
